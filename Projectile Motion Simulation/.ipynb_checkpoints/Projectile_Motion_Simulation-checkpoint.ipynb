# Simulation of Projectile Motion with Wind Influence

This project simulates the motion of a projectile launched from the ground towards the north, incorporating the effects of gravity and air resistance, as well as the influence of wind from different directions. The simulation utilizes Euler's method to numerically solve the equations of motion and generates both 2D and 3D visualizations of the projectile's trajectory.

## Project Components

### 1. Importing Libraries
We begin by importing the necessary libraries:
- `matplotlib`: For plotting and visualizations.
- `numpy`: For numerical operations.
- `mpl_toolkits.mplot3d`: For 3D plotting.
- `IPython.display`: For updating plots in Jupyter notebooks.

### 2. ProjectileTrajectory Class
The core of the project is the `ProjectileTrajectory` class, which models the trajectory of the projectile. It contains the following components:

#### a. Initialization (`__init__` method)
The class is initialized with several parameters:
- `mass`: Mass of the projectile (kg).
- `initial_speed`: Initial speed of the projectile (m/s).
- `angle`: Launch angle in degrees.
- `diameter`: Diameter of the projectile (m).
- `time_step`: Time step for the simulation.
- `wind_speed`: Speed of the wind (default is 0).
- `gravity`: Gravitational constant (default is 9.8 m/s²).
- `drag_coeff`: Drag coefficient (default is 0.47 for a sphere).
- `air_density`: Density of air (default is 1.2 kg/m³).
- `wind_dir`: Direction of the wind (1 - S to N, 2 - E to W, 3 - N to S, 4 - W to E).

#### b. Calculating Trajectory (`get_trajectory` method)
This method computes the trajectory of the projectile using Euler's method. It:
- Initializes velocity and position components.
- Computes drag force due to initial speed and wind.
- Updates acceleration components based on wind direction.
- Iteratively updates the position and velocity of the projectile until it hits the ground.

#### c. 2D Animation (`animation_2d` method)
This method generates a 2D animation of the projectile's trajectory:
- Iterates through each time step.
- Updates the plot every 4th frame or on the last frame.
- Clears the previous output and pauses briefly to create an animation effect.

#### d. 3D Animation (`animation_3d` method)
This method generates a 3D animation of the projectile's trajectory:
- Iterates through each time step.

```python
import numpy as np
from IPython import display
import matplotlib.pyplot as plt
```

```python
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

```python
# Projectile fired from the ground towards the north
class ProjectileTrajectory():
    
    def __init__(self, mass, initial_speed, angle, diameter, time_step, wind_speed=0, gravity=9.8, drag_coeff=0.47, air_density=1.2, wind_dir=1):
        self.mass = mass  # Mass of the projectile [kg]
        self.gravity = gravity  # Gravitational constant [m/s^2]
        self.initial_speed = initial_speed  # Initial speed [m/s]
        self.angle = angle  # Launch angle in degrees [degrees]
        self.drag_coeff = drag_coeff  # Drag coefficient (for a sphere)
        self.diameter = diameter  # Diameter of the projectile cross-section (sphere) [m]
        self.air_density = air_density  # Air density [kg/m^3]
        self.time_step = time_step  # Time step length
        self.wind_dir = wind_dir  # Wind direction: 1 - S-->N, 2 - E-->W, 3 - N-->S, 4 - W-->E
        self.wind_speed = wind_speed  # Wind speed
    
    def get_trajectory(self):
        
        # Velocity components
        vel_y = [self.initial_speed * np.cos(self.angle / 180 * np.pi)]
        vel_z = [self.initial_speed * np.sin(self.angle / 180 * np.pi)]
        vel_x = [0]
        
        # Projectile position at time t = 0 (launch moment)
        pos_x = [0]  # Lateral displacement (due to wind)
        pos_y = [0]  # Distance traveled
        pos_z = [0]  # Height
        time = [0]  # Time steps
        
        # Drag force acting on the initial speed V0 [m*kg/s^2]
        drag_force = (1 / 2) * self.drag_coeff * (np.pi * self.diameter ** 2) * self.air_density * self.initial_speed ** 2
        
        # Drag force due to wind speed
        drag_force_wind = (1 / 2) * self.drag_coeff * (np.pi * self.diameter ** 2) * self.air_density * self.wind_speed ** 2 
        
        # Acceleration (deceleration) for the velocity components Vx and Vy [m/s^2]  
        if self.wind_dir == 1:
            acc_x = [0]  
            acc_y = [(-drag_force * np.cos(self.angle / 180 * np.pi) + drag_force_wind) / self.mass] 
        elif self.wind_dir == 2:
            acc_x = [-drag_force_wind / self.mass]  
            acc_y = [-(drag_force * np.cos(self.angle / 180 * np.pi)) / self.mass] 
        elif self.wind_dir == 3:
            acc_x = [0]  
            acc_y = [(-drag_force * np.cos(self.angle / 180 * np.pi) - drag_force_wind) / self.mass] 
        elif self.wind_dir == 4:
            acc_x = [drag_force_wind / self.mass]  
            acc_y = [-(drag_force * np.cos(self.angle / 180 * np.pi)) / self.mass]   
            
        acc_z = [-self.gravity - (drag_force * np.sin(self.angle / 180 * np.pi) / self.mass)]
    
        # Euler's method to compute variables for subsequent time steps

        # Loop runs until the projectile hits the ground
        i = 0  
        while pos_z[i] >= 0:     
            time.append(time[i] + self.time_step)  # Add next time step
    
            # Velocity reduced by drag force in the next time step 
            vel_x.append(vel_x[i] + self.time_step * acc_x[i])
            vel_y.append(vel_y[i] + self.time_step * acc_y[i])
            vel_z.append(vel_z[i] + self.time_step * acc_z[i])
    
            # Projectile position in the next time step
            pos_x.append(pos_x[i] + self.time_step * vel_x[i])
            pos_y.append(pos_y[i] + self.time_step * vel_y[i])    
            pos_z.append(pos_z[i] + self.time_step * vel_z[i]) 
    
            # New velocity V computed from new (drag-reduced) velocity components Vy and Vz
            speed = np.sqrt(vel_y[i + 1] ** 2 + vel_z[i + 1] ** 2)
            # New drag force and new accelerations for the new velocity
            drag_force = (1 / 2) * self.drag_coeff * (np.pi * self.diameter ** 2) * self.air_density * speed ** 2
            drag_force_wind = (1 / 2) * self.drag_coeff * (np.pi * self.diameter ** 2) * self.air_density * vel_x[i + 1] ** 2 
            if self.wind_dir == 1:
                acc_x.append(0) 
                acc_y.append((-drag_force * np.cos(self.angle / 180 * np.pi) + drag_force_wind) / self.mass)
            elif self.wind_dir == 2:
                acc_x.append(-drag_force_wind / self.mass)
                acc_y.append(-(drag_force * np.cos(self.angle / 180 * np.pi)) / self.mass)
            elif self.wind_dir == 3:
                acc_x.append(0)
                acc_y.append((-drag_force * np.cos(self.angle / 180 * np.pi) - drag_force_wind) / self.mass)
            elif self.wind_dir == 4:
                acc_x.append(drag_force_wind / self.mass)
                acc_y.append(-(drag_force * np.cos(self.angle / 180 * np.pi)) / self.mass)  
                
            acc_z.append(-self.gravity - (drag_force * np.sin(self.angle / 180 * np.pi) / self.mass))
    
            i += 1
            
        return pos_x, pos_y, pos_z, time

    
    def animation_2d(self, pos_y, pos_z, time):
        for t in range(len(time)):
            if ((t % 4) == 0) or (t == len(time)):
                fig, ax = plt.subplots(figsize=(5, 5), dpi=60)
                ax.scatter(pos_y[t], pos_z[t], color="gold")
                ax.set_xlim(-0.5, max(pos_y) + 1)
                ax.set_ylim(0, max(pos_z) + 1)
                display.clear_output(wait=True)
                plt.pause(0.00001)
                
    def animation_3d(self, pos_x, pos_y, pos_z, time):
        for t in range(len(time)):
            if ((t % 4) == 0) or (t == len(time)):
                fig = plt.figure(figsize=(5, 5), dpi=100)
                ax = fig.add_subplot(projection='3d')
                ax.scatter(pos_x[t], pos_y[t], pos_z[t], color="gold")
                ax.set_xlim3d(-1, max(pos_x))
                ax.set_ylim3d(0, max(pos_y))
                ax.set_zlim3d(0, max(pos_z))
                ax.set_xlabel('x')
                ax.set_ylabel('y')
                ax.set_zlabel('z')
                display.clear_output(wait=True)
                plt.pause(0.00001)
```

## Running the Simulation
To run the simulation:
1. Create an instance of the `ProjectileTrajectory` class with the desired parameters.
2. Call the `get_trajectory` method to compute the trajectory.
3. Use the `animation_2d` or `animation_3d` methods to visualize the trajectory.

## Example

```python
# Create a projectile instance
projectile = ProjectileTrajectory(mass=1, initial_speed=50, angle=45, diameter=0.1, time_step=0.01, wind_speed=5, wind_dir=1)
```

```python
# Get the trajectory
pos_x, pos_y, pos_z, time = projectile.get_trajectory()
```

```python
# Generate 2D animation
projectile.animation_2d(pos_y, pos_z, time)
```

```python
# Generate 3D animation
projectile.animation_3d(pos_x, pos_y, pos_z, time)
```
